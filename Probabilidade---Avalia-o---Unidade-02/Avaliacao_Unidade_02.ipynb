````python
print("2. (3 pontos) Realize um estudo envolvendo as propriedades de assimétria e curtose para as turmas IMD0033 (probabilidade) consolidadas em 2017.1. Utilize gráficos e explique os resultados. Reforce também os horários das disciplinas nos resultados. ")
    
_24M12 = matricula_componente_2017_1[matricula_componente_2017_1.id_turma == 57597912]
_26N34 = matricula_componente_2017_1[matricula_componente_2017_1.id_turma == 57597913]

_24M12 = _24M12.dropna()
_26N34 = _26N34.dropna()

_24M12_unique = _24M12.pivot_table(index = "id_discente", values = "media_final", aggfunc = np.mean)
_24M12_unique = _24M12_unique.reset_index()
_24M12_unique.to_csv(path_or_buf="datasets_gerados/Probabilidade_24M12.csv")

_26N34_unique = _26N34.pivot_table(index = "id_discente", values = "media_final", aggfunc = np.mean)
_26N34_unique = _26N34_unique.reset_index()
_26N34_unique.to_csv(path_or_buf="datasets_gerados/Probabilidade_26N34.csv")

fig, (ax1, ax2) = plt.subplots(figsize=(10,5), ncols=2, nrows=1)

fig.tight_layout()

plt.suptitle("IMD0033 (Probabilidade)", fontsize=20, y = 1.09)

ax1.set_title("24M12", y = 1, fontsize=12)
ax2.set_title("26N34", y = 1, fontsize=12)

# Plot the mean in red
ax1.axvline(_24M12_unique['media_final'].mean(), color="r")
ax2.axvline(_26N34_unique['media_final'].mean(), color="r")

# Plot the median in green
ax1.axvline(_24M12_unique['media_final'].median(), color="g")
ax2.axvline(_26N34_unique['media_final'].median(), color="g")

sns.distplot(_24M12_unique['media_final'], ax=ax1, bins=20)
sns.distplot(_26N34_unique['media_final'], ax=ax2, bins=20)

print('\nAlgumas informações relevantes :',
        '\n24M12 curtose: ', kurtosis(_24M12_unique['media_final']),
        '\n26N34 curtose: ', kurtosis(_26N34_unique['media_final']),
        '\n24M12 média: ', _24M12_unique['media_final'].mean(),
        '\n26N34 média: ', _26N34_unique['media_final'].mean(),
        '\n24M12 mediana: ', _24M12_unique['media_final'].median(),
        '\n26N34 mediana: ', _26N34_unique['media_final'].median())
  
plt.show()
````

Com base na curtose podemos perceber uma curva platicúrtica em relação a distribuição normal, em ambos os gráficos das turmas. Pode-se perceber também que o desempenho na turma ofertada pela manhã é um pouco melhor do que na da noite.
