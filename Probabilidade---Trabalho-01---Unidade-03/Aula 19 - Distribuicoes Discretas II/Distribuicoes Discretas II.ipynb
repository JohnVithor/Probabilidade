```python
import matplotlib.pyplot as plt
import math
import seaborn as sns
from scipy import linspace
from scipy.stats import binom
from scipy.stats import poisson
from scipy.stats import hypergeom
sns.set()
```

```python
fig1 = plt.figure( figsize = ( 15, 10 ) )

fig1.add_subplot( 2, 2, 1 )
outcome_counts = linspace( 0, 10, 11 )
outcome_probs = poisson.pmf( outcome_counts, 1 )
plt.bar( outcome_counts, outcome_probs, label = 'lambda = 1' )
plt.legend( loc = 'best' )

fig1.add_subplot(2,2,2)
outcome_counts = linspace(0,10,11)
outcome_probs = poisson.pmf(outcome_counts,4)
plt.bar(outcome_counts, outcome_probs, label = 'lambda = 4' )
plt.legend( loc = 'best' )

fig1.add_subplot(2,2,3)
outcome_counts = linspace(0,10,11)
outcome_probs = poisson.pmf(outcome_counts,10)
plt.bar(outcome_counts, outcome_probs, label = 'lambda = 10' )
plt.legend( loc = 'best' )

plt.show()
```

```python
fig2 = plt.figure( figsize = ( 15, 10 ) )

fig2.add_subplot( 2, 2, 1 )
outcome_counts = linspace( 0, 10, 11 )
outcome_probs = poisson.cdf( outcome_counts, 1 )
plt.bar(outcome_counts, outcome_probs, label = 'lambda = 1' )
plt.legend( loc = 'best' )


fig2.add_subplot( 2, 2, 2 )
outcome_counts = linspace( 0, 10, 11 )
outcome_probs = poisson.cdf( outcome_counts, 4 )
plt.bar(outcome_counts, outcome_probs, label = 'lambda = 4' )
plt.legend( loc = 'best' )

fig2.add_subplot( 2, 2, 3 )
outcome_counts = linspace( 0, 10, 11 )
outcome_probs = poisson.cdf( outcome_counts, 10 )
plt.bar(outcome_counts, outcome_probs, label = 'lambda = 10' )
plt.legend( loc = 'best' )

plt.show()
```

```python
outcome_counts = linspace( 0, 10, 11 )

X = binom.pmf( outcome_counts, 20, .05 )
Y = poisson.pmf( outcome_counts, 20*.05 )

plt.scatter( outcome_counts, X, color = 'r', label = 'X~binomial(20,0.05)', linewidths = 7)
plt.scatter( outcome_counts, Y, color = 'b', label = 'Y~poisson(20*0.05)' , linewidths = 7)

plt.legend( loc = 'best' )

plt.title( 'Distribuição Poisson - Aproximação', weight = 'bold', size = 15 )
plt.show()
```

```python
outcome_counts = linspace( 0, 20, 21 )

X = hypergeom.pmf( outcome_counts, 100, 5, 4 )

plt.scatter( outcome_counts, X, color = 'b', label = 'X~hypergeom(k,100,5,4)', linewidths = 9 )

plt.legend( loc = 'upper right' )

plt.title( 'Distribuição Hipergeométrica - pmf', weight = 'bold', size = 17 )
plt.show()
```

```python
outcome_counts = linspace( 0, 20, 21 )

X = hypergeom.cdf( outcome_counts, 100, 5, 4 )

plt.scatter( outcome_counts, X, color = 'b', label = 'X~hypergeom(k,100,5,4)', linewidths = 9)

plt.legend( loc = 'lower right' )

plt.title( 'Distribuição Hipergeométrica - cdf', weight = 'bold', size = 17 )
plt.show()
```
